```python
import cv2
from matplotlib import pyplot as plt
import os
import numpy as np
import mediapipe as mp
import pandas as pd
```

```python
hands_train = np.load(os.path.join("/Users/illusivesheep/Repositories/ультра датасет", "npy", "train_coords.npy"), allow_pickle="true")
hands_test = np.load(os.path.join("/Users/illusivesheep/Repositories/ультра датасет", "npy", "test_coords.npy"), allow_pickle="true")
hands_val = np.load(os.path.join("/Users/illusivesheep/Repositories/ультра датасет", "npy", "val_coords.npy"), allow_pickle="true")
```

```python
path = "/Users/illusivesheep/Repositories/ультра датасет/train/"
classes_folders = [folder for folder in os.listdir(path) if "." not in folder]
modes = ["train", "test"]
```

```python
df = pd.DataFrame()
gesture = {"камень": 0, "бумага": 1, "ножницы": 2, "коза": 3, "не_нравится": 4, "класс": 5}
try:
    os.mkdir(os.path.join(path, "dots"))
except FileExistsError:
    pass
for folder in classes_folders:
    
    try:
        os.mkdir(os.path.join(path, "dots", folder))
    except FileExistsError:
        pass
    
    folder_path = os.path.join(path, folder)
    images_paths = [image for image in os.listdir(folder_path) if ".png" in image]
    
    for image_path in images_paths:
        image = cv2.imread(os.path.join(folder_path, image_path))
        mp_hands = mp.solutions.hands
        with mp_hands.Hands(
                static_image_mode=True,
                max_num_hands=1,
                min_detection_confidence=0.7) as hands:
            results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    
                    coef_x = 0.5-hand_landmarks.landmark[9].x
                    coef_y = 0.5-hand_landmarks.landmark[9].y
                    blank_image = np.zeros((224,224,3), np.uint8)
                    
                    for i, point in enumerate(hand_landmarks.landmark):
                        X = int((coef_x + point.x ) * 224)
                        Y = int((coef_y + point.y ) * 224)
                        cv2.circle(blank_image, (X, Y), 1, (255,255,255), 3)
                        
                        cv2.imwrite(os.path.join(path, "dots", folder, image_path), blank_image)
                        df = df.append({ 'Image': image_path,
                                         'landmark_id': i,
                                         'X': X, 
                                         'Y': Y,
                                         'Z': point.z,
                                         'class': gesture[folder]}, ignore_index=True)


df[["class", "landmark_id", "X", "Y"]] = df[["class", "landmark_id", "X", "Y"]].astype(int)
df.to_csv(os.path.join(path, "dots", "train_dots.csv"))
```

```python
aggregated_df = df[["Image", "X", "Y", "class"]].copy().groupby(['Image'], as_index=False).agg([min, max]).reset_index()
# aggregated_df = pd.pivot_table(df, index="Image", columns=["landmark_id"], values=["X", "Y"], aggfunc=[max, min])
# df_new["max"] = aggregated_df.X["max"].copy
aggregated_df.reset_index(inplace=True, level = 0)
aggregated_df["X_min"] = aggregated_df.X["min"]
aggregated_df["X_max"] = aggregated_df.X["max"]
aggregated_df["Y_min"] = aggregated_df.Y["min"]
aggregated_df["Y_max"] = aggregated_df.Y["max"]
aggregated_df["class_Id"] = aggregated_df["class"]["max"]
aggregated_df = aggregated_df[["Image", "X_min", "X_max", "Y_min", "Y_max", "class_Id"]]
aggregated_df = aggregated_df.reset_index()
aggregated_df
```

```python
def generate_images(x):

    gesture = {0: "rock", 1: "paper", 2: "scissors", 3: "goat", 4: "dislike", 5: "like"}
    image = cv2.imread(os.path.join("/Users/illusivesheep/Repositories/ультра датасет/images_train_test", 
                                    "train",
                                    gesture[x["class_Id"]], 
                                    x["Image"]))
    image = image[x.Y_min:x.Y_max, x.X_min:x.X_max]
    cv2.imwrite(os.path.join("/Users/illusivesheep/Repositories/ультра датасет/images_train_test", 
                             "train", 
                             "crop", 
                             x.Image), image)
    
aggregated_df.apply(generate_images, axis=1)
```
